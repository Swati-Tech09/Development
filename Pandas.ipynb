```python
import os
import numpy as np
import pandas as pd
```

```python
os.getcwd()
os.chdir('D:\Data')
Emp_Ds=pd.read_csv("Emp.csv")
Emp_Ds.shape
```

```python
Emp_Ds.head(10)
```

```python
# Group BY Aggregate functions, based on Role 
df_groupby_role = Emp_Ds.groupby(['Role'])
df_groupby_role
# select only required columns, which you want to print
df_groupby_role = df_groupby_role[["Role", "Salary"]]
# get the average
df_groupby_role_mean = df_groupby_role.mean()
print(df_groupby_role_mean)
```

```python
df_salary_by_role = Emp_Ds.groupby(['Role'])[["Role", "Salary"]].sum().reset_index()
print(df_salary_by_role)
```

```python
##How to copy a dataframe using selected columns?
Col1=Emp_Ds[['Role','Dept']]
#Pass deep = True for deep copy
df=Col1.copy(deep='True')
df
```

```python
#Rename the existing DataFrame's columns
df.rename(columns={'Role': 'JobRole', 'Dept': 'OrgId'}, inplace=True)
df
```

```python
#Create a subset of dataframe
df_Female = Emp_Ds[(Emp_Ds['Sex'] == 'F')].reset_index()
df_Female
```

```python
# Create a dataset of Employee belongs to Dept 10, 20
Emp_dept = Emp_Ds[Emp_Ds["Dept"].isin([10,20])]
Emp_dept
```

```python
# Create or combine 2 dataframe into 1
Emp_Ds1=pd.read_csv("Emp1.csv")
Emp_Ds1.shape
Emp_Ds1
```

```python
result=Emp_Ds.append(Emp_Ds1)
result

# using ignore_index
res = pd.concat([Emp_Ds, Emp_Ds1], ignore_index=True)
 
res
```

```python
#Both pivot_table and groupby are used to aggregate your dataframe. The difference is only with regard to the shape of the result

pd.pivot_table(Emp_Ds1, index=['Role'], columns=['Dept'], values=["Salary"], aggfunc=np.mean, fill_value=0)


```
